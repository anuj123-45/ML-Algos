```python
import pandas as pd
import matplotlib.pyplot  as plt
```

```python
df=pd.read_csv('height_weight.csv')
df.shape
```

```python
# hist plot
plt.hist(df.height,bins=20,rwidth=0.8)
plt.xlabel('height')
plt.ylabel('count')
plt.show()
```

```python
from scipy.stats import norm
import numpy as np
plt.hist(df.height, bins=20, rwidth=0.8, density=True)
plt.xlabel('Height')
plt.ylabel('Count')

rng = np.arange(df.height.min(), df.height.max(), 0.1)
plt.plot(rng, norm.pdf(rng,df.height.mean(),df.height.std()))
```

```python
plt.hist(df.weight,bins=20,rwidth=0.8)
plt.xlabel('weight')
plt.ylabel('count')
plt.show()
```

```python
from scipy.stats import norm
import numpy as np
plt.hist(df.weight, bins=20, rwidth=0.8, density=True)
plt.xlabel('Weight')
plt.ylabel('Count')

rng = np.arange(df.weight.min(), df.weight.max(), 0.1)
plt.plot(rng, norm.pdf(rng,df.weight.mean(),df.weight.std()))
```

```python
#Outliers removal using height
q1=df.height.quantile(0.25)
q3=df.height.quantile(0.75)
IQR=q3-q1
```

```python
IQR
```

```python
upper_limit=q3+1.5*IQR
lower_limit=q1-1.5*IQR
```

```python
lower_limit,upper_limit
```

```python
df[(df.height<lower_limit)|(df.height>upper_limit)]
```

```python
#Outliers removal using weight
Q1=df.weight.quantile(0.25)
Q3=df.weight.quantile(0.75)
IQR=Q3-Q1
```

```python
upper_limit=Q3+1.5*IQR
lower_limit=Q1-1.5*IQR
```

```python
upper_limit,lower_limit
```

```python
df[(df.weight<lower_limit)|(df.weight>upper_limit)]
```
