```python
df=pd.DataFrame(digits.data,columns=digits.feature_names)
```

```python
df.head()
```

```python
df.describe()
```

```python
X=df
y=digits.target
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_Scaled=scaler.fit_transform(X)
X_Scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_Scaled,y,test_size=0.2,random_state=45)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
X.head(3)
```

```python
y
```

```python
from sklearn.decomposition import PCA
pca=PCA(0.95)
X_pca=pca.fit_transform(X)
X_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
X_train_pca,X_test_pca,y_train_pca,y_test_pca=train_test_split(X_pca,y,test_size=0.2,random_state=45)
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(X_train_pca,y_train_pca)
model.score(X_test_pca,y_test_pca)*100
```

```python
pca=PCA(n_components=2)
```

```python
X_pca=pca.fit_transform(X)
X_pca
```

```python
pca.explained_variance_ratio_
```

```python
X_train_pca,X_test_pca,y_train_pca,y_test_pca=train_test_split(X_pca,y,test_size=0.2,random_state=45)
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(X_train_pca,y_train_pca)
model.score(X_test_pca,y_test_pca)*100
```
